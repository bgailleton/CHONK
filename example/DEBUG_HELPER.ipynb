```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML

%matplotlib widget
```

```python

dfs = []
n_dt = 4
for i in range(1,n_dt + 1):
    tdf = pd.read_csv("debug_lake_%s.csv" %(i) ).sort_values("pitnode")
    tdf = tdf[tdf.active == 1]
    dfs.append(tdf)
```

```python
# display(HTML(dfs[-1].to_html()))
dfs[-1]["delta_Vwm"] = dfs[-1]["Vw"] - dfs[-1]["Vwmax"]
tdf = dfs[-1]#[dfs[-1]["Vw_outlet"]>0]
HTML(tdf.to_html())
```
