```python
import numpy as np
#import xsimlab as xs
#import xarray as xr
import matplotlib.pyplot as plt
```

```python
dt = 0.01
dx = 0.01

results = []
ndt = int(1e6) #nb jusqu'au quel on va, not the range though!!!

break_point = 5000

x = np.arange(1, 100, dx)
z = np.sin(x)
KD = np.zeros_like(z)
KD[break_point:] = 2e-3
KD[:break_point] = 1e-3
```

```python
fig,ax = plt.subplots()
ax.plot(x,z)
ax2 = ax.twinx()
ax2.plot(x,KD, color = 'red')
ax.set_xlabel("distance (m)")
ax.set_ylabel("elevation (m)")
```

```python
F = (dt/dx)
```

```python
F
```

```python
for i in range(ndt):
    z[1:-1] = ((z[2:] - z[1:-1])/dx * KD[1:-1] - ((z[1:-1]-z[:-2])/dx * KD[:-2])) * F + z[1:-1]
    z[-1] = z[-2]
    if(i%5000 == 0): #"if the remainder from the division i/500 = 0”, it will be true once every 500 iteration of i, 
        #it is very useful when you want to save outputs from a model every x steps
        results.append(z.copy()) #if I don't do this, I don't save any values and I can't plot the diffusion like this
       # results.append(KD.copy())
```

```python
fig,ax = plt.subplots()
ax.plot(x,z)
for res in results:
    ax.plot(x,res)
ax.set_xlabel("distance (m)")
ax.set_ylabel("elevation (m)")
```
